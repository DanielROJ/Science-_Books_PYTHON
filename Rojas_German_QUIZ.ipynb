# Punto 1 (Solución)

#### Linea de Codigo  (No 1)

Del Dataframe Original selecciona unas columnas especificas, para construir un nuevo Dataframe mas indicado.

```python
movie4 = movie[['movie_title','title_year','content_rating','budget']]  
movie4.head() ##Print Result
```

## ----------------------------------------------------------------------------

#### Linea de Codigo  (No 2)

Del Sub-Dataframe obtenido anteriormente, organiza la siguientes Columnas de forma decendiente:

* 'title_year'
* 'content_rating'

Por el contrario, organiza de forma ascendente la siguiente Columna:

* 'budget'

```python
movie4_sorted = movie4.sort_values(['title_year','content_rating','budget'],
                                  ascending=[False,False,True])
movie4_sorted.head()## Print Result 
```

## ----------------------------------------------------------------------------

#### Linea de Codigo  (No 3)

Del Dataframe organizado anteriormente, se borran registros o datos repetidos de las Columnas:

* 'title_year'
* 'content_rating'

Esto provoca, que por cada dato eliminado de la columna se borre la fila de datos  que lo contenia.
Por ultimo se imprimen las **10** primeras filas del Dataframe

```python
movie4_sorted.drop_duplicates(subset=['title_year','content_rating']).head(10)
```
